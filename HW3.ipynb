# 1.

First, we show that it is consistent for all $\alpha$. The corresponding characteristic polynomials are $\rho(z) = z^2 +\alpha z -(1+\alpha)$ and $\sigma(z) = \frac{4+3\alpha}{2} z - \frac{\alpha}{2} z^2$. We check that $\rho(1) = 0$ by computing

$$\rho(1) = 1^2 +\alpha  -(1+\alpha)= 0.$$

Then we need to check that $\rho'(1) = \sigma(1)$. So we compute

$$\rho'(1) = 2 +\alpha,$$
$$\sigma(1) = \frac{4+3\alpha}{2}  - \frac{\alpha}{2} = 2 +\alpha,$$

which means that any $\alpha$ should satisfy these conditions. Thus the method is consistent for any $\alpha$.

Now we find the range of zero stability. We factor $\rho(z) = z^2 +\alpha z -(1+\alpha) = (z-1)(z+\alpha+1)$. This means our roots are 1 and $-\alpha-1$. We want their absolute values to both be less than or equal to 1(as long as they are not repeated). This means $|\alpha+1| \leq 1$, which implies $-1 \leq \alpha+1 \leq 1$, from which we get $-2 \leq \alpha \leq 0$. We notice that if $\alpha = -2$, we have a repeated root, so we exclude that one. Thus, the range of zero stability is $-2 < \alpha \leq 0$.

We now show that it is of the second order. We compute the following Taylor expansions:

$$ x(t_{k+1}) = x(t_k) + h x'(t_k) + \frac12 h^2 x''(t_k) + \dots $$
$$ x(t_{k+2}) = x(t_k) + 2h x'(t_k) + 2 h^2 x''(t_k) + \dots $$
$$ x'(t_{k+1}) = x'(t_k) + h x''(t_k) + \frac12 h^2 x'''(t_k) + \dots $$
$$ x'(t_{k+2}) = x'(t_k) + 2h x''(t_k) + 2 h^2 x'''(t_k) + \dots $$

Now we compute the truncation error:

$$\tau_k = \frac{x(t_k) + 2h x'(t_k) + 2 h^2 x''(t_k) + \dots + \alpha(x(t_k) + h x'(t_k) + \frac12 h^2 x''(t_k) + \dots) - (1+\alpha) x(t_k)}{h}$$
$$+ \frac{\alpha}{2}(x'(t_k) + 2h x''(t_k) + 2 h^2 x'''(t_k) + \dots) - \frac{4+3\alpha}{2}(x'(t_k) + h x''(t_k) + \frac12 h^2 x'''(t_k) + \dots)$$
$$ = \frac{x_k + 2h x_k' + 2 h^2 x_k'' + \alpha x_k + \alpha h x_k' + \frac{\alpha}2 h^2 x_k'' - x_k - \alpha x_k  + \dots}{h} 
+ \frac{\alpha}{2}(x_k' + 2h x_k'' + 2 h^2 x_k''') - \frac{4+3\alpha}{2}(x_k' + h x_k'' + \frac12 h^2 x_k''') + \dots$$
$$ = \frac{2h x_k' + 2 h^2 x_k'' + \alpha h x_k' + \frac{\alpha}2 h^2 x_k''}{h} 
+ \frac{\alpha}{2}(x_k' + 2h x_k'' + 2 h^2 x_k''') - \frac{4+3\alpha}{2}(x_k' + h x_k'' + \frac12 h^2 x_k''') + \dots $$
$$ = (2 + \alpha) x_k' + (\frac{\alpha h}2 + 2 h) x_k''
+ \frac{\alpha}{2}x_k' + \alpha h x_k'' + \alpha h^2 x_k''' - \frac{4+3\alpha}{2} x_k' - \frac{4+3\alpha}{2} h x_k'' - \frac{4+3\alpha}{4}  h^2 x_k''') + \dots $$
$$ = (2 + \alpha) x_k' + \frac{\alpha}{2}x_k' - \frac{4+3\alpha}{2} x_k'
+ (\frac{\alpha h}2 + 2 h) x_k'' + \alpha h x_k'' - \frac{4+3\alpha}{2} h x_k''
+ \alpha h^2 x_k''' - \frac{4+3\alpha}{4}  h^2 x_k''') + \dots $$
$$ = \frac{4+3\alpha}{2} x_k' - \frac{4+3\alpha}{2} x_k'
+ \frac{4+3\alpha}{2} h x_k'' - \frac{4+3\alpha}{2} h x_k''
+ \alpha h^2 x_k''' - \frac{4+3\alpha}{4}  h^2 x_k''') + \dots $$
$$ = \alpha h^2 x_k''' - \frac{4+3\alpha}{4}  h^2 x_k''') + \dots 
= O(h^2).$$
Thus, it is of the second order for any $\alpha$.

Lastly, we show the region of A-stability. We write the function $\Pi(x,z) = \rho(x) -z\sigma(x) = x^2 +\alpha x -(1+\alpha)- z( \frac{4+3\alpha}{2} x - \frac{\alpha}{2} x^2)$. We can group for $x$ to get
$$ x^2(1 +  \frac{z\alpha}{2}) + x(\alpha - \frac{4+3\alpha}{2}) -(1+\alpha), $$
from which we can apply the quadratic formula to get

$$x = \frac{\frac{4z+3z\alpha}{2}-\alpha \pm \sqrt{z^2(\frac94 \alpha^2 + 6\alpha + 4) - z\alpha(\alpha+2) + (\alpha+2)^2}}{2+z\alpha}.$$

So now I just have to find the values of $\alpha$ for which the absolute value of that thing is less than or equal to 1 and the real part of $z$ is negative. From zero stability $\alpha$ must be between 0 and -2, so we could just test that.

# b. 
Now we code the Forward Euler Method for this problem. 

```python
# INPUT: initial time a, ending time b, interval length as h,
#        initial point (x1, x2) 
# OUTPUT: two arrays approximating the corresponding components

def fem(a, b, h, x1, x2):
    c1=[x1]
    c2=[x2]
    for i in range(int((b-a)/h)):
        t1 = (-33.4*x1 + 66.6*x2)
        t2 = (33.3*x1 - 66.7*x2)       
        x1 += h * t1
        x2 += h * t2
        c1.append(x1)
        c2.append(x2)
        a += h
    return c1,c2
```

For this part, our time step is 0.1. So we run and plot it accordingly.

```python
bsol = fem(0,3,0.1,3,0)
x = np.linspace(0, 3.0, num=31)
plt.plot(x, bsol[0], label='First Component')
plt.plot(x, bsol[1], label='Second Component')
plt.legend(loc='upper right')
```

# a.

We write code for the Euler forward method.

```python
# INPUT: number of mesh points n+2, time step k,
#        number of time steps M 
# OUTPUT: u(1,x) approximated by the given inputs

def heat_forward(n,k,M):
    h = 1/(n+1)
    s = k/(h**2)
    v = np.linspace(-n-1, n+1, n+2)
    w = np.exp(-100*(v*h)**2)
    for j in range(1, M+1):
        v[0] = 0
        v[-1] = 0
        for i in range(1,len(v)-1):
            v[i] = s*w[i-1] + (1 - 2*s)*w[i] + s*w[i+1]
        w[1:len(w)-1] = v[1:len(w)-1]
    return w    
```

Then we plot it.

```python
plt.plot(np.linspace(-1,1,51), heat_forward(49,0.1,9))
```

Now we do the backward Euler, but first we copy over the tridiagonal solver from 104B, HW6.

```python
# INPUT: nxn tridiagonal, positive definite matrix A and nx1 vector d, from the system Ax=d
# NOTE: we used numpy arrays and to guarantee accuracy,
#       write A_11 as a decimal
# OUTPUT: the solution vector x
def tri_solver(n,a,d,c,b):
    l = np.zeros(n-1)
    m = np.zeros(n)
    
    if d[0] == 0:
            return "Error: matrix is not positive definite."
    m[0] = d[0]
    for j in range(n-1):
        l[j] = a[j] / m[j]
        m[j+1] = d[j+1] - l[j] * c[j]
        if m[j+1] == 0:
            return "Error: matrix is not positive definite"
       
    # forward substitution Ly = d
    y = np.zeros(n)        
    y[0] = b[0]
    for j in range(1,n):
        y[j] = b[j] - l[j-1] * y[j-1]
  
    # backward substitution Ux = y
    x = np.zeros(n)
    x[n-1] = y[n-1] / m[n-1]
    for j in range(n-2, -1, -1):
        x[j] = (y[j] - c[j] * x[j+1]) / m[j]
    return x
```

Now we do the actual backward method.

```python
# INPUT: number of mesh points n, time step k,
#        number of time steps M 
# OUTPUT: u(1,x) approximated by the given inputs

def heat_backward(n,k,M):
    h = 1/(n+1)
    s = k/(h**2)
    v = np.linspace(-n, n, n)
    w = np.exp(-100*(v*h)**2)
    c = -np.ones(n-1)*s
    a = -np.ones(n-1)*s
    for j in range(1, M+1):
        d = np.ones(n) + 2*s 
        w = tri_solver(n,a,d,c,w)
    return w
```

Lastly we plot it.

```python
plt.plot(np.linspace(-1,1,49), heat_backward(49,0.1,10))
```

Now we do the Trapezodial Method.

```python
# INPUT: number of mesh points n+2, time step k,
#        number of time steps M 
# OUTPUT: u(1,x) approximated by the given inputs

def heat_crank(n,k,M):
    h = 1/(n+1)
    s = k/(h**2)
    v = np.linspace(-n, n, n)
    w = np.exp(-100*(v*h)**2)
    c = -np.ones(n-1)*s/2
    a = -np.ones(n-1)*s/2
    for j in range(1, M+1):
        d = np.ones(n) + s 
        for i in range(1,len(v)-1):
            v[i] = s*w[i-1]/2 + (1 - s)*w[i] + s*w[i+1]/2
        w[1:len(w)-1] = v[1:len(w)-1]
        w = tri_solver(n,a,d,c,w)
    return w
```

And then we plot it.

```python
plt.plot(np.linspace(-1,1,49), heat_crank(49,0.1,10))
```

# e.

We compare and explain now.

For Forward Euler, only the last graph(in part d) looked normal. This is because for that one we have $\frac{0.0001}{0.02^2} = 0.25 < 0.5$, so it is stable. For the other graphs before that, we get some weird oscillations because we don't have stability. Our time step gives us ratios greater than $\frac12$.

Meanwhile, for the Backward Euler method, our solution is always "OK" because this method is always stable.

Although the Crank(Trapezodial) method is always stable, we need $\frac{k}{h^2}$ to be less than or equal to 1/2 for the maximum mesh errors to be kept in check. That's why we see similar problems as with the forward euler with the earlier graphs.
